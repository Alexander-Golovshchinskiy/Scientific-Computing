```python
import matplotlib.pyplot as plt
import numpy as np
# Time points and absorbance values
time_points = [0, 15, 30, 45, 60, 75, 90, 105, 120]
absorbance_values = [0.344, 0.369, 0.402, 0.445, 0.454, 0.577, 0.656, 0.735, 0.807]


log_absorbance_values = [np.log10(x) for x in absorbance_values]
# Plotting the growth curve
plt.figure(figsize=(8, 6))
plt.plot(time_points, log_absorbance_values, marker='o', linestyle='-', color = 'black')
plt.xlabel('Time (min)')
plt.ylabel('log($A_{600}$)')
# plt.ylim(0)
plt.xticks(range(0, 121, 15))
plt.grid(False)
plt.savefig('growth_curve.png')
plt.show()
print(log_absorbance_values)
```
